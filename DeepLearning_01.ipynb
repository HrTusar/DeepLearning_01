```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= np.arange(-5,5,0.01)  #-5=start,5=end,0.01=step
```

```python
def plot(func, yaxis = (-1.4, 1.4)):
    plt.ylim(yaxis)   #ylimite=-1.4 to 1.4
    plt.locator_params(nbins = 5) #params create bins,bar
    plt.xticks(fontsize = 14)
    plt.yticks(fontsize = 14) #ticks Control fontsize
    plt.axhline(lw = 1, c = 'black') #axhline=Horizontal line
    plt.axvline(lw = 1, c = 'black') #axvline= vertical line
    plt.grid(alpha = 0.4, ls = '-.')
    plt.box(on = None) 
    plt.plot(x, func(x), c = 'r', lw = 3) # Recursion = Repeated 
```

```python
binary_step = np.vectorize(lambda x: 1 if x > 0 else 0, otypes = [np.float])
```

```python
plot(binary_step, yaxis = (-0.4, 1.4)) #define constructor
```

```python
plot(binary_step) #defult constructor without define
```

```python
piecewise_linear = np.vectorize(lambda x: 1 if x>3 else 0 if x<3 else 2) #use multiple condition
```

```python
plot(piecewise_linear)
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))  #sigmoid function use, exp= exponential, sigmoid value must be 0 to 1 range
```

```python
plot(sigmoid)
```

```python
relu = np.vectorize(lambda x: x if x>0 else 0, otypes = [np.float]) # relu function,as like as Regression
```

```python
plot(relu)
```

```python
leaky_relu = np.vectorize(lambda x: x if max(0.1 *x, x) else 0, otypes = [np.float])
# Exponential Linear Units [ELU]
```

```python
plot(leaky_relu)
```

```python
elu = np.vectorize(lambda x: x if x > 0 else 0.5* (np.exp(x) - 1), otypes = [np.float])
```

```python
plot(elu)
```

```python
def softplus(x):
  return np.log(1+ np.exp(x))
```

```python
plot(softplus)
```
